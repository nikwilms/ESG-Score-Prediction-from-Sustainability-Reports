```python
import bs4 as bs
import requests
import pickle
import datetime as dt
import yfinance as yf
import pandas as pd
import json
import urllib.request
import yesg
```

```python
resource = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = bs.BeautifulSoup(resource.text, 'html.parser')
table = soup.find('table', {'class': 'wikitable sortable'})
```

```python
tickers = []

for row in table.findAll('tr')[1:]:
    ticker = row.findAll('td')[0].text
    tickers.append(ticker)
```

```python
tickers = [s.replace('\n', '') for s in tickers]
```

```python
tickers
```

```python
import yesg

dataframes = []
for ticker in tickers:
    try:
        df = pd.DataFrame(yesg.get_historic_esg(ticker))
        df['Company_Symbol'] = ticker
        dataframes.append(df)
    except:
        pass

df = pd.concat(dataframes)
df['timestamp'] = df.index
df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
        
```

```python
df.dropna(inplace=True)
```

```python
df['Company_Symbol'].nunique()
```

```python
df['timestamp'] = df.index
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
```

```python
df['year'] = df['timestamp'].dt.year
```

```python
df
```

```python
cleaned_df = df.groupby(['year', 'Company_Sembol']).mean()
```

```python
cleaned_df.to_csv('SP500_EGS_Score_avarage_per_year.csv')
```
